```python
import pandas as pd
import numpy as np

df = pd.read_csv('/content/plan_estudio_bigquery.csv')
```

```python
# prompt: quisiera poder tener activo los links de la columna Recursos para abrir el respectivo link

def make_clickable(url):
    return f'<a href="{url}" target="_blank">{url}</a>'

# Ensure the 'Recurso' column exists and contains strings
if 'Recurso' in df.columns:
    # Apply the function only to the 'Recurso' column
    df_styled = df.style.format({'Recurso': make_clickable})
    # Display the styled dataframe as HTML
    from IPython.display import display, HTML
    display(HTML(df_styled.to_html(escape=False)))
else:
    print("The 'Recurso' column was not found in the DataFrame.")
```
