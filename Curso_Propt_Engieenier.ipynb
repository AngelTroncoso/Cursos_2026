```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np # Import numpy for handling potential non-numeric correlation results

def eda_automatico(archivo_csv):
    """
    Realiza un Análisis Exploratorio de Datos (EDA) automático en un archivo CSV,
    genera gráficos y proporciona recomendaciones.

    Args:
        archivo_csv (str): La ruta al archivo CSV.
    """

    try:
        df = pd.read_csv(archivo_csv)
    except FileNotFoundError:
        print(f"Error: El archivo '{archivo_csv}' no se encontró.")
        return
    except pd.errors.EmptyDataError:
        print(f"Error: El archivo '{archivo_csv}' está vacío.")
        return
    except pd.errors.ParserError:
        print(f"Error: No se pudo analizar el archivo '{archivo_csv}'.  Verifica el formato CSV.")
        return

    print("\n--- Información General del Dataset ---")
    df.info()

    print("\n--- Primeras 5 Filas del Dataset ---")
    print(df.head())

    print("\n--- Estadísticas Descriptivas ---")
    print(df.describe())

    print("\n--- Valores Faltantes ---")
    print(df.isnull().sum())

    # --- Visualizaciones ---

    # 1. Distribución de Variables Numéricas
    print("\n--- Generando Histogramas para Variables Numéricas ---")
    numericas = df.select_dtypes(include=['number']).columns
    for columna in numericas:
        plt.figure(figsize=(8, 6))
        sns.histplot(df[columna], kde=True)
        plt.title(f'Distribución de {columna}')
        plt.xlabel(columna)
        plt.ylabel('Frecuencia')
        plt.show()

    # 2. Distribución de Variables Categóricas
    print("\n--- Generando Gráficos de Barras para Variables Categóricas ---")
    categoricas = df.select_dtypes(include=['object']).columns
    for columna in categoricas:
        plt.figure(figsize=(8, 6))
        sns.countplot(x=df[columna])
        plt.title(f'Distribución de {columna}')
        plt.xlabel(columna)
        plt.ylabel('Cantidad')
        plt.xticks(rotation=45, ha='right')  # Rotar etiquetas si son largas
        plt.tight_layout() # Ajustar el layout para evitar que las etiquetas se superpongan
        plt.show()

    # 3. Relaciones entre Variables Numéricas (Matriz de Correlación)
    print("\n--- Generando Matriz de Correlación ---")
    # Select only numerical columns for correlation calculation
    df_numeric = df.select_dtypes(include=[np.number])
    # Check if there are enough numerical columns to calculate correlation
    if df_numeric.shape[1] > 1:
        plt.figure(figsize=(10, 8))
        sns.heatmap(df_numeric.corr(), annot=True, cmap='coolwarm')
        plt.title('Matriz de Correlación')
        plt.show()
    else:
        print("No hay suficientes columnas numéricas para generar una matriz de correlación.")

    # 4. Relación entre Variables Numéricas (Scatter Plots)
    print("\n--- Generando Scatter Plots para Variables Numéricas ---")
    numericas = df_numeric.columns # Use the columns from the numeric DataFrame
    for i in range(len(numericas)):
        for j in range(i + 1, len(numericas)):
            plt.figure(figsize=(8, 6))
            sns.scatterplot(x=df[numericas[i]], y=df[numericas[j]])
            plt.title(f'{numericas[i]} vs {numericas[j]}')
            plt.xlabel(numericas[i])
            plt.ylabel(numericas[j])
            plt.show()
    if len(numericas) <= 1:
        print("No hay suficientes columnas numéricas para generar scatter plots.")


    # 5. Relación entre Variables Categóricas y Numéricas (Box Plots)
    print("\n--- Generando Box Plots para Variables Categóricas vs Numéricas ---")
    for columna_cat in categoricas:
        for columna_num in numericas: # Use the numerical columns identified earlier
            plt.figure(figsize=(8, 6))
            sns.boxplot(x=df[columna_cat], y=df[columna_num])
            plt.title(f'{columna_num} por {columna_cat}')
            plt.xlabel(columna_cat)
            plt.ylabel(columna_num)
            plt.xticks(rotation=45, ha='right')
            plt.tight_layout()
            plt.show()
    if not categoricas.any() or not numericas.any():
        print("No hay suficientes columnas categóricas o numéricas para generar box plots.")


    # --- Recomendaciones ---
    print("\n--- Recomendaciones ---")
    print("- **Manejo de Valores Faltantes:**  Si hay valores faltantes, considera imputarlos (media, mediana, moda) o eliminarlos, dependiendo del porcentaje y la importancia de la variable.")
    print("- **Outliers:**  Identifica y trata los outliers.  Pueden ser errores de datos o valores legítimos pero extremos.  Considera transformaciones (logarítmica, raíz cuadrada) o técnicas de capping.")
    print("- **Ingeniería de Características:**  Crea nuevas variables a partir de las existentes para mejorar el poder predictivo de los modelos.")
    print("- **Escalado de Características:**  Escala las variables numéricas (estandarización, normalización) si planeas usar algoritmos sensibles a la escala.")
    print("- **Análisis de Correlación:**  Investiga las variables altamente correlacionadas.  Considera eliminar una de ellas si son redundantes.")
    print("- **Análisis de Variables Categóricas:**  Considera la codificación one-hot o label encoding para variables categóricas.")
    print("- **Validación Cruzada:**  Utiliza la validación cruzada para evaluar el rendimiento de los modelos de manera robusta.")
    print("- **Selección de Características:**  Utiliza técnicas de selección de características para identificar las variables más relevantes.")


# --- Ejemplo de Uso ---
archivo_csv = '/content/What-did-Chile-export-in-2023-.csv'  # Reemplaza con la ruta a tu archivo CSV
eda_automatico(archivo_csv)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# 1. Leer archivo CSV (ajusta el nombre si lo deseas)
archivo_csv = "/content/What-did-Chile-export-in-2023-.csv"  # Cambia por tu archivo
df = pd.read_csv(archivo_csv)

# Crear carpeta para guardar los gráficos
carpeta_salida = "graficos_powerbi"
os.makedirs(carpeta_salida, exist_ok=True)

# 2. Estadísticas generales
print(df.describe(include='all'))

# 3. Histograma para cada variable numérica
for columna in df.select_dtypes(include=['float64', 'int64']).columns:
    plt.figure(figsize=(8, 5))
    sns.histplot(df[columna].dropna(), kde=True, color='skyblue')
    plt.title(f'Distribución de {columna}')
    plt.tight_layout()
    plt.savefig(f'{carpeta_salida}/hist_{columna}.png')
    plt.close()

# 4. Gráficos de correlación
plt.figure(figsize=(12, 8))
correlacion = df.corr(numeric_only=True)
sns.heatmap(correlacion, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Mapa de calor de correlaciones')
plt.tight_layout()
plt.savefig(f'{carpeta_salida}/heatmap_correlacion.png')
plt.close()

# 5. Boxplots de variables numéricas por cada columna categórica
categoricas = df.select_dtypes(include=['object', 'category']).columns
numericas = df.select_dtypes(include=['float64', 'int64']).columns

for cat in categoricas:
    for num in numericas:
        if df[cat].nunique() < 10:  # Para evitar muchos grupos
            plt.figure(figsize=(10, 6))
            sns.boxplot(data=df, x=cat, y=num, palette='Set2')
            plt.title(f'{num} por {cat}')
            plt.xticks(rotation=45)
            plt.tight_layout()
            plt.savefig(f'{carpeta_salida}/boxplot_{num}_por_{cat}.png')
            plt.close()

print("✅ Gráficos generados y guardados en la carpeta:", carpeta_salida)
```

```python
#carga de modelo de traduccion
from transformers import MarianMTModel, MarianTokenizer # Corrected module and class names
# Import MarianMTModel and MarianTokenizer from their specific submodules
# The correct class name is MarianMTModel, not MarianMTMmodel
# from transformers.models.marian.modeling_marian import MarianMTModel
# from transformers.models.marian.tokenization_marian import MarianTokenizer


model_name = "Helsinki-NLP/opus-mt-en-es"

# Use the corrected class names
tokenizer = MarianTokenizer.from_pretrained(model_name)
model = MarianMTModel.from_pretrained(model_name)
```

```python
#Usamos el modelo para generar traducción
#preparar el texto a traducir

texto_ingles = ['hello,how are you?', 'this is a translation test.']
tokens = tokenizer(texto_ingles, return_tensors='pt', padding=True, truncation=True)
```

```python
#generara la traducción
traducción_tokens = model.generate(**tokens)
#decodificamos los tokens traducidos para obtener el texto final
traduccion_texto = tokenizer.batch_decode(traducción_tokens, skip_special_tokens=True)
```

```python
#ver el resultado
for i, t in zip( texto_ingles, traduccion_texto):
  print(f'ingles: {i} -> español: {t}')
```
